```python
numbers = [int(n) for n in input("Введіть числа: ").split()]

sorted_num = sorted(numbers, reverse=True)

def non_prime(n):
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return True
    return False

non_prime_numbers = [n for n in sorted_num if non_prime(n)]

print(non_prime_numbers)
```
